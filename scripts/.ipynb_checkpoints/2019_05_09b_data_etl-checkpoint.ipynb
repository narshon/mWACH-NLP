## I bet the `Study ID` column in `df1` corresponds to the `pid` column in `df2`

```python
df1['Study ID'].nunique()
```

```python
df2.pid.nunique()
```

```python
set(df1['Study ID']) - set(df2.pid)
```

```python
df = pd.merge(df1, df2, left_on='Study ID', right_on='pid')
df
```

## Hacky way to group the messages by system message

```python
df['system_msg'] = (df.sent_by == 'system')
print('Number of system messages sent:', df.system_msg.sum())
```

```python
df['msg_group'] = df.system_msg.cumsum()
df.msg_group.value_counts()
```

```python
y = (df.msg_group.value_counts() > 1)
```

```python
y.tail(8)
```

```python
print('Number of system messages with a "response":', y.sum())
```

## Hacky way to form feature vectors

```python
t = df[df.sent_by == 'system']
t = t.set_index('msg_group')
t.shape
```

```python
t.index.nunique()
```

```python
t['response'] = y
t.response.tail(8)
```

```python
t.head()
```

```python
t.language.unique()
```

```python
t['swahili'] = t.language == 'swahili'
t['english'] = t.language == 'english'
t['luo'] = t.language == 'luo'
```

```python
X = t.filter(['edd_wk', 'chars', 'words', 'swahili', 'english', 'luo']).astype(float)
y = t.response.astype(float)
```

```python
(100 * t.groupby('language').response.mean().sort_values(ascending=False)).plot.bar(color='grey')
plt.xticks(rotation=0)
plt.ylabel('Engagement Rate (%)')
```

## simple prediction from sklearn

```python
import sklearn.linear_model
```

```python
clf = sklearn.linear_model.LogisticRegression(solver='lbfgs')
clf.fit(X, y)
```

```python
pr = clf.predict_proba(X)
plt.hist(pr[:,0])
plt.xlabel('Predicted Probability')
plt.ylabel('Number of messages')
plt.axis(xmin=0, xmax=1)
plt.grid();
```

```python
import sklearn.ensemble
```

```python
clf = sklearn.ensemble.GradientBoostingClassifier()
clf.fit(X, y)
```

```python
pr = clf.predict_proba(X)
plt.hist(pr[:,0])
plt.xlabel('Predicted Probability')
plt.ylabel('Number of messages')
plt.axis(xmin=0, xmax=1)
plt.grid();
```
