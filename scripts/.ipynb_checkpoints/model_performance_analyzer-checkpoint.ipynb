# Model Performance Analyzer

We review several models with respect to their performance outputs for a simple classification problem: 
Given a message sent to a participant, what is the probability of a getting back a response? 
The following models have been tested in this notebook and their accuracy and roc_auc_scores recorded and plotted for comparisons.
    1. Logistic regression using known features (edd_wk,study_wk, past_response,native_langauge,delivery_status,swahili, english,luo)
    2. Gradient Boosting classifier using known features.
    3. Logistic regression using bow only
    4. Logistic regression using bow and known features
    5. Gradient Boosting classifier using bow only
    6. Gradient Boosting classifier using bow, and known features
    7. Logistic regression using bi-grams and known features
    8. Gradient Boosting classifier using bi-grams and known features
    9. Gradient boosting using word2vec embeddings and known features
    10. Gradient boosting using doc2vec embeddings and known features

```python
#We need to know number of responses each system message received. --- we will come back later to remove nurse messages
messages_df['msg_group'] = messages_df.system_msg.cumsum()
messages_df.msg_group.value_counts()
```

```python
#Our responses vector is all those messages with a value count of more than 1
response = (messages_df.msg_group.value_counts() > 1)
```

```python
systemMessagesDF = messages_df[messages_df.sent_by == "system"]
systemMessagesDF = systemMessagesDF.set_index('msg_group')
systemMessagesDF['response'] = response
systemMessagesDF
```

```python
%time
#calculate past response rate of system message (did this participant respond to the last message?)
#let loop through the messages and check current participant has a previous system message with a response
#for mid, pid, has_response from systemMesaagesDF:
def past_response(mid2, pid2):
    has_prev_response = 0
    #get prev system message to this participant
    # last system message from this participant before the current one
    last_msg = systemMessagesDF[(systemMessagesDF.pid == pid2) & (systemMessagesDF.mid < mid2) ].tail(1) 
    
    if(len(last_msg) > 0):
        return last_msg.iloc[0]['response']
    else:
        return False
    
    
#past_response(808, 15)
systemMessagesDF['past_response'] = systemMessagesDF.apply(lambda row: past_response(row.mid, row['pid']), axis=1)
```

# 1. Logistic regression using features of the message only

```python
X = systemMessagesDF.filter(features_list).astype(float).values
y = systemMessagesDF.response.astype(float).values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,  random_state=5555)
```

```python
#We now fit the model
logreg = sklearn.linear_model.LogisticRegression(solver='lbfgs', max_iter=600)
logreg.fit(X_train, y_train)
```

```python
#Let's explore the coefficients
print(logreg.coef_)
```

```python
pr = logreg.predict_proba(X_test)
predict = logreg.predict(X_test)
plt.hist(pr[:,0])
plt.xlabel('Predicted Probability')
plt.ylabel('Number of messages')
plt.axis(xmin=0, xmax=1)
plt.grid();
```

```python
#Testing the performance of this model.
acc_ = accuracy_score(y_test, predict)
acc_
```

```python
#Let's also do roc-auc-score
roc_ = roc_auc_score(y_test, pr[:,1])
roc_
```

```python
print(classification_report(y_test, predict))
```

```python
# Let's see the results visually
fpr, tpr, thresholds = roc_curve(y_test, pr[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.4f)' % roc_)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

```python
#K-fold cross validation
rkf = RepeatedKFold(n_splits=5, n_repeats=2, random_state=2652124)
accuracy = []
roc = []
count = 0
acc = 0
model = sklearn.linear_model.LogisticRegression(solver='lbfgs')
for train_index, test_index in rkf.split(X):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    p = model.predict_proba(X_test)
    accuracy.append(accuracy_score(y_test, pred)*100)  #converting the preditions to 100%
    roc.append(roc_auc_score(y_test, p[:,1]))
kfold_acc= statistics.mean(accuracy)
kfold_roc= statistics.mean(roc)
kfold_roc
```

```python
#Get highest predictor and lowest predictor from the model
coeficients = model.coef_
feature_imp_dict = dict(zip(features_list, coeficients[0]))
feature_imp_dict_sorted = list(sorted(feature_imp_dict.items(), key=lambda kv: kv[1]))
lowest_predictor = list(feature_imp_dict_sorted[0:3])
highest_predictor = list(feature_imp_dict_sorted[len(feature_imp_dict_sorted)-3:])
h_pred = []
for var in highest_predictor:
    h_pred.append(var[0])
l_pred = []
for var in lowest_predictor:
    l_pred.append(var[0])
print(feature_imp_dict)
```

```python
# Let's AUC after KFold
fpr, tpr, thresholds = roc_curve(y_test, p[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.4f)' % kfold_roc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

```python
#Save results in a dataframe
model1 = pd.DataFrame({'Model':'LogReg', 
                        'Features': ",".join(features_list),
                        'Acc_score': acc_,
                        'roc_auc_score': roc_,
                        'kfold_cv_acc': kfold_acc,
                        'kfold_cv_roc': kfold_roc,
                        'highest_predictor': ",".join(h_pred),
                        'lowest_predictor': ",".join(l_pred),
                       }, index = ['Model1'] )
model1
```

```python
def avg_word_vectors(wordlist,size):
    """
    returns a vector of zero for reviews containing words where none of them
    met the min_count or were not seen in the training set
    
    Otherwise return an average of the embeddings vectors
    
    """
    
    sumvec=np.zeros(shape=(1,size))
    wordcnt=0
    
    for w in wordlist:
        if w in model:
            sumvec += model[w]
            wordcnt +=1
    
    if wordcnt ==0:
        return sumvec
    
    else:
        return sumvec / wordcnt
```

```python
#K-fold cross validation
rkf = RepeatedKFold(n_splits=5, n_repeats=2, random_state=2652124)
accuracy = []
roc = []
count = 0
acc = 0
GB_model = sklearn.ensemble.GradientBoostingClassifier()
for train_index, test_index in rkf.split(X_all_, y):
    X_train, X_test = X_all_[train_index], X_all_[test_index]
    y_train, y_test = y[train_index], y[test_index]
    GB_model.fit(X_train, y[train_index])
    pred = GB_model.predict(X_test)
    p = GB_model.predict_proba(X_test)
    accuracy.append(accuracy_score(y_test, pred)*100)  #converting the preditions to 100%
    roc.append(roc_auc_score(y_test, p[:,1]))

kfold_acc_ = statistics.mean(accuracy)
kfold_roc_ = statistics.mean(roc)
kfold_roc_
```
